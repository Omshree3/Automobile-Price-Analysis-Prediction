```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Automobile_data.csv')
df.info()
```

```python
df.columns
```

```python
df.sample(5)
```

```python
df['normalized-losses'].unique()
# (df['normalized-losses'] == '?').sum()
# df['normalized-losses'].isna().sum()
```

```python
# cleaning normalized-losses

df['normalized-losses'].replace('?', np.nan, inplace = True) # rplacing ? with nan
df['normalized-losses']= df['normalized-losses'].astype(dtype='float')
df['normalized-losses'].info()
```

```python
df['normalized-losses'] = df.groupby('make')['normalized-losses'].transform(lambda x: x.fillna(x.mean()))
df['normalized-losses'].isna().sum() # still na is there 
```

```python
# where it seems those brands dont have normalized losses 

df[df['normalized-losses'].isna()]
```

```python
df['normalized-losses'] = df['normalized-losses'].fillna(df['normalized-losses'].mean())
df['normalized-losses'] = df['normalized-losses'].apply(lambda x: round(x))
df['normalized-losses'].astype(dtype='int')
df['normalized-losses'].unique()
```

```python
df.sample(5)
```

```python
(df['bore']=='?').sum()
```

```python
# df['bore'].unique()
# df['num-of-doors'].replace('?', 'four', inplace=True)
df[df['bore'] == '?']
```

```python
df.loc[(df['make'] == 'mazda') & (df['body-style'] == 'hatchback')]
```

```python
df['bore'].replace('?', np.nan, inplace = True)
df['bore']= df['bore'].astype('float')
df['stroke'].replace('?', np.nan, inplace = True)
df['stroke'] = df['stroke'].astype('float')
```

```python
cond = (df['make'] == 'mazda') & (df['body-style'] == 'hatchback')

df.loc[cond, 'bore'] = df.loc[cond, 'bore'].fillna(df.loc[cond, 'bore'].mean())

df.loc[cond, 'stroke'] = df.loc[cond, 'stroke'].fillna(df.loc[cond, 'stroke'].mean())

# now bore and stroke column is also clean
```

```python
df['horsepower'].unique()
```

```python
(df['horsepower']=='?').sum()
```

```python
df['horsepower'].replace('?', np.nan, inplace=True)
df['horsepower'].isna().sum()
```

```python
df['horsepower']= df['horsepower'].astype(dtype='float')
df['horsepower'].fillna(df['horsepower'].median(), inplace=True)
df['horsepower']= df['horsepower'].astype(dtype='int')
```

```python
df['peak-rpm'].unique()
```

```python
(df['peak-rpm']=='?').sum()
```

```python
df.loc[df['peak-rpm'] == '?', :]
```

```python
df['peak-rpm'].replace('?', np.nan, inplace= True)
```

```python
df['peak-rpm']= df['peak-rpm'].astype('float')
df['peak-rpm'].fillna(df['peak-rpm'].median(), inplace= True)
df['peak-rpm']= df['peak-rpm'].astype('int')
```

```python
df['peak-rpm'].isna().sum()
```

```python
df['price'].unique()
```

```python
(df['price']=='?').sum()
```

```python
df.loc[(df['price']== '?'), :]
```

```python
df['price'].replace('?', np.nan, inplace=True)
df['price'] = df['price'].astype(dtype= 'float')
df['price'] = df.groupby(['make', 'engine-size'])['price'].transform(lambda x: x.fillna((x.median())))
df['price'].isna().sum()

```

```python
df[df['price'].isna()]
```

```python
df['price'] = df.groupby(['body-style', 'engine-size'])['price'].transform(lambda x: x.fillna((x.median())))
```

```python
cond = df['make'] == 'porsche'
df.loc[cond, 'price'] = df.loc[cond, 'price'].fillna(df.loc[cond, 'price'].median())
df['price']= df['price'].astype(dtype='int')
```
